```python
import pandas as pd
```

```python
df = pd.read_csv('air_quality.csv')
```

```python
print(df.head(10))
```

```python
print(df.info())
```

```python
max_temp = df['Temp'].max()
print("\nLa temperatura màxima és:", max_temp)
```

```python
max_temp_row = df[df['Temp'] == max_temp]
max_temp_month = max_temp_row['Month'].values[0]
max_temp_day = max_temp_row['Day'].values[0]
print(f"\nLa temperatura màxima es va registrar el mes {max_temp_month} i el dia {max_temp_day}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x='Wind', y='Solar.R', data=df)
plt.title("Relació entre el vent i el Solar.R")
plt.xlabel("Vent")
plt.ylabel("Solar.R")
plt.show()

print("No sembla haver-hi una clara correlació entre el vent i el Solar.R.")
```

```python
# 7. Gràfic de barres: temperatura mitjana per mes
mean_temp_by_month = df.groupby('Month')['Temp'].mean()
mean_temp_by_month.plot(kind='bar', color='skyblue')
plt.title("Temperatura mitjana per mes")
plt.xlabel("Mes")
plt.ylabel("Temperatura mitjana")
plt.show()

print("La temperatura mitjana augmenta progressivament amb els mesos, indicant que les dades són probablement d'una estació càlida.")
```

```python
# 8. Gràfic de punts: relació entre l’ozó i la temperatura
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Ozone', y='Temp', data=df)
plt.title("Relació entre l’ozó i la temperatura")
plt.xlabel("Ozó")
plt.ylabel("Temperatura")
plt.show()

print("Hi ha una tendència a l'alça, suggerint que a major concentració d'ozó, major temperatura.")
```
